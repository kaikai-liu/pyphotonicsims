## MZI resonator with balaced detection
 - Coupling power
 $$\kappa = \sin{\frac{\pi \delta n L_c}{\lambda}}$$
 - Coupling rate for resonator
 $$\gamma_{ex} = \frac{c\log{1/(1-\kappa^2)}}{n_g L}$$


```python
r_in = 2
r_ex = 1
ka2 = 0.5

phi_list = np.array([0, 1/4, 2/4, 3/4, 4/4])*np.pi
dw = np.linspace(-20, 20, 200)
for phi in phi_list:
    y, y1, y2 = PDB_MZI_resonator(dw, r_in, r_ex, ka2, ka2, phi)
    plt.plot(dw, y, label = f'{phi/np.pi:.2f}$\pi$')

plt.axhline(y = 0, color = 'tab:gray', linestyle='-')
plt.xlim((-20, 20))
plt.ylim((-1.1, 1.1))
plt.legend()
plt.xlabel(r'$\Delta\nu$ (MHz)')
plt.ylabel(r'$\Delta T = T_1 - T_2$')
```
