## Waveguide mode simulation on x $\mu$m by 80 nm

```python
base_path = 'Simulations/lumerical mode/waveguide mode/11um by 40(80)nm'
filename = 'neff_data_6umby80nm_sweep_w.mat'
filepath = get_box_path(base_path, filename)
matdata = loadmat(filepath)
w_list = matdata['w_list'].T
neff = matdata['neff']
neff = neff[:, [0, 1, 3, 2, 4]]

mode_list = ['TE0', 'TE1', 'TE2', 'TM0', 'TM1']

plt.plot(w_list*1e6, neff)
plt.xlim((0, 12))
plt.ylim((1.445, 1.467))
plt.xlabel(r'Waveguide width ($\mu$m)')
plt.ylabel(r'n$_{neff}$')
plt.legend(mode_list)
```
